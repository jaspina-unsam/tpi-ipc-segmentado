## Unión y limpieza de archivos `.parquet` provenientes de R, de las EPH en formato onda 1996-2003

```python
df_list = []
_msg = "Archivos agregados a la lista: "
for rds_file in sorted(os.listdir('data/')):
    if rds_file.lower().endswith('.rds'):
        df = pyreadr.read_r(os.path.join('data/', rds_file))[None]
        _msg += f"{rds_file.replace('.rds', '')}, "
        df_list.append(df)

print(_msg)
```

```python
base_full = pd.concat(df_list)
```
